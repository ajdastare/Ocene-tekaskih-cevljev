```python
# naložimo paket pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
modeli = pd.read_csv('obdelani-podatki/vsi-modeli.csv', index_col='ocena')

# ker bomo delali z velikimi razpredelnicami, povemo, da naj se vedno izpiše le 20 vrstic
pd.options.display.max_rows = 20
```

```python
#za vsako znamko, seštejemo število modelov 
modeli.groupby('znamka').size().sort_values(ascending=False).plot.bar()
```

Vidimo, da ima največje število modelov tekaških čevljev Asics. Sledita ji znamki Nike in Adidas.

```python
modeli2 = pd.read_csv('obdelani-podatki/vsi-modeli.csv')
modeli_z_o = modeli2[['znamka','ocena']]
modeli_z_o.groupby('znamka').mean().sort_values('ocena',ascending=False).plot.bar()
#povprečna ocena po znamkah
```

Najboljšo povprečno oceno ima 361 Degrees, Jordan in Skechers. Prej omenjene znamke pa si sledijo v tem vrstnem redu: Adidas,Nike, Asics. 

```python
m3 = modeli2[['znamka','stevilo ocen']]
m3.groupby('znamka').mean().sort_values('stevilo ocen', ascending = False).plot.bar()
```

Največ ocen ima Salomon, Vibram  FiveFingers in Nike.
